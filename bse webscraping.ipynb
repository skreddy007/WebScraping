```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
from urllib.request import Request, urlopen

req = Request('https://www.bseindia.com/markets/equity/EQReports/StockPrcHistori.aspx?expandable=7&scripcode=500209&flag=sp&Submit=G', headers={'User-Agent': 'Microsoft Edge'})
webpage = urlopen(req).read()
webpage
```

```python
soup=BeautifulSoup(webpage,'lxml')
soup
```

```python
tableheaderdata=soup.find_all('td',{'class':'innertable_header1'})   #find the headers of table using class name
tableheaderdata
    
```

```python
l=[]
for data in tableheaderdata:
    i=data.get_text() #extract each header from table
    import string
    i.translate({ord(c): None for c in string.whitespace})
    l.append(i)
l.remove('* Spread')
l1=[]
l1.append(l)

```

```python
df=pd.DataFrame(l1)
df
```

```python
tabledata=soup.find_all('tr',{'class':"TTRow"})
tabledata
```

```python
l2=[]
for i in tabledata:
    data=i.find_all('td')
    
    l3=[]
    for j in data:
        l3.append(j.get_text())
    l2.append(l3)
l2 
```

```python
df1=pd.DataFrame(l2)
df1
```

```python
df2=pd.concat([df,df1])
df2
```

```python
df3=df2.rename(columns=df2.iloc[0])
df4=df3.drop(df3.index[0])
df4
```

```python
df4.to_csv('Infy Sensex.csv',index=False)
```
